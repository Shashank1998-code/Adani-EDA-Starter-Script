```python
ticker=["ADANIENT.NS","ADANIPOWER.NS","ADANIGREEN.NS","AWL.NS","ADANIPORTS.NS","ATGL.NS","ADANITRANS.NS"]
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(15,9))
for i in ticker:
    y=yf.Ticker(i)
    plt.plot(y.dividends)
    plt.legend("{}".format(i))
    plt.grid()
```
